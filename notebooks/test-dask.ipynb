## Some simple data discovery computations on the GB catalog

We look for the event with minimum frequency. 

```python
f = h5py.File("/home/maude/data/LDC/sangria/1.7/sangria.h5", 'r')
gbs = f["sky/dgb/cat"]
```

```python
freqs = da.from_array(gbs['Frequency'], chunks=(100000,1))
freqs
```

```python
min_freq = freqs.min()
min_freq
```

```python
%%time
min_freq.compute()
```

```python
min_freq.visualize()
```
